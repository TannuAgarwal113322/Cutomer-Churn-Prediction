### **Importing Required Modules**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

### **Reading CSV File**

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.sample(5)
```

### **Analysis**

```python
df.info()
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges']=df['TotalCharges'].astype(np.float64)
```

```python
df.info()
```

### **Checking for NULL Value**

```python
df.isna().sum()
```

### **Handling NULL Value**

```python
df=df.dropna()
```

```python
df
```

### **Encoding**

**Ordinal Encoding**

```python
ord=OrdinalEncoder(categories=[['No', 'DSL', 'Fiber optic'],['Month-to-month', 'One year', 'Two year']], dtype=np.int32)
```

```python
df[['InternetService','Contract']]=ord.fit_transform(df[['InternetService','Contract']])
```

```python
df
```

**Label Encoding**

```python
lbl=LabelEncoder()
```

```python
df['Churn']=lbl.fit_transform(df['Churn'])
```

```python
df
```

```python
df
```

```python
df.info()
```

```python
cols=list()
for col in df.columns:
    if(df[col].dtype=='O'):
        cols.append(col)

cols.remove('customerID')
```

```python
cols
```

**One Hot Encoding**

```python
df=pd.get_dummies(df, columns=cols)
df=df.replace({True:1, False:0})
```

```python
df.sample(5)
```

```python
churn=df.pop('Churn')
df.insert(len(df.columns), 'Churn', churn)
```

```python
customerId=df.pop('customerID')
```

```python
df
```

### **Diviede it into Input and Output Column**

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
```

### **Train Test Split**

```python
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.25, random_state=0)
```

### **Scaling**

```python
scaler=MinMaxScaler()
```

```python
X_train=pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test=pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
```

```python
X_test
```

```python
X_train.to_csv('X_train.csv', index=False)
Y_train.to_csv('Y_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
Y_test.to_csv('Y_test.csv', index=False)
```
