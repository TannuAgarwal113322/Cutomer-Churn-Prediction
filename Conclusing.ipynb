```python
import matplotlib.pyplot as plt
```

```python
accuracy={'SVM Linear': 0.8071672354948806, 'SVM Polynomial': 0.79806598407281, 'Naive Bayes': 0.6899886234357224}
precision={'SVM Linear': 0.6508379888268156, 'SVM Polynomial': 0.6352941176470588, 'Naive Bayes': 0.4427570093457944}
recall={'SVM Linear':  0.5212527964205816, 'SVM Polynomial': 0.48322147651006714, 'Naive Bayes': 0.8478747203579419}
f1={'SVM Linear': 0.5788819875776398, 'SVM Polynomial': 0.5489199491740788, 'Naive Bayes': 0.5817344589409056}
```

```python
accuracy_cat=accuracy.keys()
accuracy_val=accuracy.values()

precision_cat=precision.keys()
precision_val=precision.values()

recall_cat=recall.keys()
recall_val=recall.values()

f1_cat=f1.keys()
f1_val=f1.values()
```

```python
plt.plot(accuracy_cat, accuracy_val, label='Accuracy', marker='o', color='red')
plt.plot(precision_cat, precision_val, label='Precision', marker='o', color='blue')
plt.plot(recall_cat, recall_val, label='Recall', marker='o', color='green')
plt.plot(f1_cat, f1_val, label='F1 Score', marker='o', color='orange')
plt.legend(loc='upper right')
plt.xlabel('Categories')
plt.ylabel('Scores')
```
